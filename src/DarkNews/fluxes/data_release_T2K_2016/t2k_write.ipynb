```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Flux (/cm^2/1E21p.o.t/50MeV)
minus = np.genfromtxt("t2kflux_2016_nd280_minus250kA.txt",unpack=True,skip_header=3)
plus = np.genfromtxt("t2kflux_2016_nd280_plus250kA.txt",unpack=True,skip_header=3)
```

```python
_, e1, e2,numu, numubar, nue , nuebar = minus 
_, e1_2, e2_2,numu_2, numubar_2, nue_2 , nuebar_2 = plus
```

```python
units = 1e-21/0.05

plt.plot(e1_2+(e2_2-e1_2)/2.0,numu_2*units,color='indigo',ls='--')
plt.plot(e1_2+(e2_2-e1_2)/2.0,numubar_2*units,color='green',ls='--')

plt.plot(e1_2+(e2_2-e1_2)/2.0,nue_2*units,color='indigo',ls='--')
plt.plot(e1_2+(e2_2-e1_2)/2.0,nuebar_2*units,color='green',ls='--')

plt.xlim(0,30)
plt.yscale('log')
```

```python
np.savetxt("../ND280_FHC.dat",np.array([e1+(e2-e1)/2.0,
                                        nue*units,numu*units,numu*units*0,nuebar*units,numubar*units,numu*units*0]).T,
          header='nus/cm^2/POT/GeV')

np.savetxt("../ND280_RHC.dat",np.array([e1+(e2-e1)/2.0,nue*units,numu*units,numu*units*0,nuebar*units,numubar*units,numu*units*0]).T,
                    header='nus/cm^2/POT/GeV')
```
