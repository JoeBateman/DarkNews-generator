```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Flux (/cm^2/POT/50MeV)
minus = np.genfromtxt("./MiniBooNE_nu_mode_flux.dat",unpack=True)
plus = np.genfromtxt("./MiniBooNE_nubar_mode_flux.dat",unpack=True)
```

```python
e1, e2,numu, numubar, nue , nuebar = minus 
e1_2, e2_2,numu_2, numubar_2, nue_2 , nuebar_2 = plus
```

```python
plt.plot(e1_2+(e2_2-e1_2)/2.0,numu_2,color='indigo',ls='--')
plt.plot(e1_2+(e2_2-e1_2)/2.0,numubar_2,color='green',ls='--')

plt.plot(e1_2+(e2_2-e1_2)/2.0,nue_2,color='indigo',ls='--')
plt.plot(e1_2+(e2_2-e1_2)/2.0,nuebar_2,color='green',ls='--')

plt.xlim(0,10)
plt.yscale('log')
```

```python
units = 1/0.05
np.savetxt("../MiniBooNE_FHC.dat",np.array([e1+(e2-e1)/2.0,
                                        nue*units,numu*units,numu*units*0,nuebar*units,numubar*units,numu*units*0]).T,
          header='nus/cm^2/POT/GeV')

np.savetxt("../MiniBooNE_RHC.dat",np.array([e1+(e2-e1)/2.0,nue*units,numu*units,numu*units*0,nuebar*units,numubar*units,numu*units*0]).T,
            header='nus/cm^2/POT/GeV')
```
