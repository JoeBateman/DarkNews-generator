```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
data_nu = np.genfromtxt("flux_dune_neutrino_ND_globes.txt")
data_nubar = np.genfromtxt("flux_dune_antineutrino_ND_globes.txt")

for i in range(1,6):
    data_nu[:,i] *= 1e-4
    data_nubar[:,i] *= 1e-4
```

```python
for i in range(1,6):
    plt.plot(data_nu[:, 0], data_nu[:,i])

plt.xlim(0.0,20)
# plt.ylim(1e-3,1e2)
plt.yscale('log')
plt.legend(frameon=False)
```
